```python
#importing all the necessary libraries 

%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python
#reading in the data from a csv file
df = pd.read_csv('RegentsData.csv')
```

```python
#checking out the beginning of the data
df.head()
```

```python
#checking out the end of the data to see if it's consistent throughout
df.tail()
```

```python
#selecting the columns I plan to use
df2 = df.iloc[:,[1,3,4,5,9]]

#reading the first 50 colums 
df2.head(50)
```

```python
#naming the colums so I can reference them at a later point
df2.columns = ['school_name', 'school_level', 'regents_exam', 'year', 'mean_score']

#checking to see that the columns are named properly
df2.head()
```

```python
#checking the number of rows before I clean the data
len(df2)
```

```python
#filtering the data to only select high school data
high_schools = df2.school_level.str.contains('High school')

df2 = df2.loc[high_schools]

#checking to see if K-8 data was filtered out
len(df2)
```

```python
#checking to see that the data was filtered out
df2
```

```python
#selecting only Geometry Regents for simplicities sake
regents_type =  df2.regents_exam.str.contains('Common Core Geometry')

df2 = df2.loc[regents_type]

#checking to see if the other regents were excluded 
len(df2)
```

```python
#getting rid of data that contains value 's' (schools with 5 students or less taking the regents)
numeric_mean = df2.mean_score.str.contains('s')

df2 = df2.loc[~numeric_mean]




```

```python
#changing the mean_score value to a numeric value instead of a string so we can use mathematical functions 
df2.mean_score = df2.mean_score.apply(pd.to_numeric)
```

```python
#checking to see if the mean_score was actually changed to a numeric data type
df2.dtypes
```

```python
#grouping the data by year
import numpy as np
group_by_year = df2.loc[:, ['school_name', 'school_level', 'regents_exam', 'year', 'mean_score']].groupby('year')
```

```python
#getting the mean of the regents score by year
avgs = group_by_year.mean()   

#changing the values we got for the mean to an array so it can be plotted 
mean_values = np.asarray(avgs)

#avgs

#setting our x and y values for our graph
x = avgs.index
y = mean_values

#creating a function to plot our data
def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.set_xlabel('Years')
    ax.plot(x, y)
    ax.margins(x=1, y=10)

```

```python
#creating a figure and subplot for our data and actually plotting our data!
fig, ax = plt.subplots()
plot(x, y, ax, 'Average Geometry Regents Scores After Common Core is Introduced', 'Average Score')
```
